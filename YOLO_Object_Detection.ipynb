```python
from ultralytics import YOLO
from google.colab import files
from IPython.display import Image, display
import os
import glob

# Load YOLO model
model = YOLO("yolov8m.pt")   # better accuracy than yolov8n

# Upload image
uploaded = files.upload()

# Get uploaded image name
image_name = list(uploaded.keys())[0]

# Run object detection
results = model(
    image_name,
    save=True,     # save framed image
    conf=0.5       # confidence threshold
)

# Find latest prediction folder automatically
predict_folders = glob.glob("runs/detect/predict*")
latest_folder = sorted(predict_folders)[-1]

# Create output image path
output_image = os.path.join(latest_folder, image_name)

# Display framed image
display(Image(filename=output_image))
```
